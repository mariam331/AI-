```python
for i in range(1,11):
    url = f"https://quotes.toscrape.com/page/{i}/"
    website = requests.get(url).text
    mywebsite = BeautifulSoup(website , "lxml")
    quotes = mywebsite.find_all("div",class_="quote")
    for quote in quotes:
        title = quote.find("span",class_="text").text
        author = quote.find("small",class_="author").text
        tags = quote.find("div",class_="tags").text
        print(title,author,tags)
```
